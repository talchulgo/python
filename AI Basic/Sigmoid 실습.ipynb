```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(a,x):
    return 1 / (1+np.exp(-a*x))
```

```python
x = np.arange(-5,5,0.1)
y = sigmoid(1,x)
y1 = sigmoid(2,x)
y2 = sigmoid(3,x)
```

```python
y
```

```python
plt.plot(x,y, label='exp(-1x)')
plt.plot(x,y1, label='exp(-2x)')
plt.plot(x,y2, label='exp(-3x)')
plt.legend()
plt.show()
```
